```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib notebook

beer_consump = pd.read_csv("istherecorrelation.csv", sep=';')
year = list(map(int, beer_consump["Year"]))
# WO_float = beer_consump["WO [x1000]"].replace(",", ".")
WO_string = [w.replace(",", ".") for w in (list(beer_consump["WO [x1000]"]))]
WO_list = list(map(float, WO_string))
WO = [w * 1000 for w in WO_list]
beer = list(map(int, beer_consump["NL Beer consumption [x1000 hectoliter]"]))

plt.figure(1)
plt.plot(WO_string, beer, 'b')
plt.xlabel("WO")
plt.ylabel("Beer consumption (x1000 hectoliter)")
plt.title("Beer consumption with respect to WO")
plt.show()
```
